```python
# ---------------------------------------------------------
# STEP 1: Saari Zaroori Libraries Import Karna
# ---------------------------------------------------------
import kagglehub
import os
import glob
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

print("Step 1: Libraries loaded!")

# ---------------------------------------------------------
# STEP 2: Dataset Download aur Paths Setup
# ---------------------------------------------------------
print("Step 2: Dataset check kar rahe hain...")
path = kagglehub.dataset_download("kmader/skin-cancer-mnist-ham10000")
csv_path = os.path.join(path, 'HAM10000_metadata.csv')
skin_df = pd.read_csv(csv_path)

# Images ka actual address (path) dhoondh kar DataFrame mein add karna
image_paths = glob.glob(os.path.join(path, '**', '*.jpg'), recursive=True)
image_path_dict = {os.path.splitext(os.path.basename(x))[0]: x for x in image_paths}
skin_df['path'] = skin_df['image_id'].map(image_path_dict)

# ---------------------------------------------------------
# STEP 3: Train-Test Split (Sirf Text Data ka)
# ---------------------------------------------------------
# RAM bachane ke liye hum images nahi, sirf unke paths ko split kar rahe hain
train_df, test_df = train_test_split(skin_df, test_size=0.20, random_state=42)
print(f"Step 3: Training ke liye {len(train_df)} aur Testing ke liye {len(test_df)} files set hain.")

# ---------------------------------------------------------
# STEP 4: Smart Data Generators (No RAM Crash)
# ---------------------------------------------------------
print("Step 4: Smart Generators start ho rahe hain...")
train_datagen = ImageDataGenerator(
    rescale=1./255,         # Scaling
    rotation_range=20,      # Augmentation
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)
test_datagen = ImageDataGenerator(rescale=1./255)

# Batch-by-Batch images load hongi (128x128 size mein)
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df, x_col="path", y_col="dx",
    target_size=(128, 128), batch_size=32, class_mode="categorical"
)
test_generator = test_datagen.flow_from_dataframe(
    dataframe=test_df, x_col="path", y_col="dx",
    target_size=(128, 128), batch_size=32, class_mode="categorical", shuffle=False
)

# ---------------------------------------------------------
# STEP 5: Class Weights Calculate Karna (Imbalance ke liye)
# ---------------------------------------------------------
# Generator ne kis bimari ko kya number diya, woh check karke weights nikal rahe hain
class_indices = train_generator.class_indices
y_train_mapped = train_df['dx'].map(class_indices)

weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_mapped),
    y=y_train_mapped
)
class_weights_dict = dict(enumerate(weights))
print("Step 5: Class weights set ho gaye taaki rare cancer miss na ho.")

# ---------------------------------------------------------
# STEP 6: Transfer Learning Pro Model (MobileNetV2)
# ---------------------------------------------------------
print("Step 6: Google MobileNetV2 'Pro' dimaag load ho raha hai...")
base_model = MobileNetV2(
    input_shape=(128, 128, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False # Purani knowledge lock kar di

tl_model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(7, activation='softmax') # 7 Cancer classes
])

tl_model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# ---------------------------------------------------------
# STEP 7: Final Training (The Magic Happens Here)
# ---------------------------------------------------------
print("\n🚀 Step 7: Final Pro Training Shuru (T4 GPU zindabad!)...")
history = tl_model.fit(
    train_generator,
    epochs=15,
    validation_data=test_generator,
    class_weight=class_weights_dict
)

print("\n🎉 Badhai ho! Pura crash-free model train ho chuka hai!")
```

# New Section

```python
import matplotlib.pyplot as plt

# 1. Figure ka size set karna (Taaki graph bada aur clear dikhe)
plt.figure(figsize=(12, 5))

# 2. Pehla Graph: ACCURACY (Sahi jawab kitne diye)
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy', color='blue', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='orange', linewidth=2)
plt.title('Model Accuracy Progress', fontsize=14)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.legend()
plt.grid(True)

# 3. Dusra Graph: LOSS (Galtiyan kitni ki)
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss', color='blue', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', color='orange', linewidth=2)
plt.title('Model Loss Progress', fontsize=14)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend()
plt.grid(True)

# Graph ko screen par dikhana
plt.tight_layout()
plt.show()
```
